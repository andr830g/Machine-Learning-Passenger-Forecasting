# Data Quality Check

```python
import pandas as pd

df_15 = pd.read_csv("data/data15.csv")
df_30 = pd.read_csv("data/data30.csv")
df_60 = pd.read_csv("data/data60.csv")
```

All time periods are present in DK local time

```python
climateHour = pd.read_csv("data/ClimateData/ClimateDataHour.csv")
```

```python
climateHour_date1_meanCloudCover = climateHour[climateHour["from"] == "2020/01/01 01:00:00"]["mean_cloud_cover"].reset_index(drop=True)[0]
df_60_date1_meanCloudCover = df_60[df_60["datetime"] == "2020-01-01 02:00:00"]["mean_cloud_cover"].reset_index(drop=True)[0]

climateHour_date1_meanCloudCover == df_60_date1_meanCloudCover  # 1 hour ahead of UTC time (winter)
```

```python
climateHour_date2_meanCloudCover = climateHour[climateHour["from"] == "2022/07/11 23:00:00"]["mean_wind_speed"].reset_index(drop=True)[0]
df_30_date2_meanCloudCov = df_30[df_30["datetime"] == "2022-07-12 01:30:00"]["mean_wind_speed"].reset_index(drop=True)[0]

climateHour_date2_meanCloudCover == df_30_date2_meanCloudCov  # 2 hour ahead of UTC time (summertime)
```

```python
climateHour_date3_meanRelativeHum = climateHour[climateHour["from"] == "2022/10/30 02:00:00"]["mean_wind_speed"].reset_index(drop=True)[0]
df_30_date3_meanRelativeHum = df_30[df_30["datetime"] == "2022-10-30 03:30:00"]["mean_wind_speed"].reset_index(drop=True)[0]

climateHour_date3_meanRelativeHum == df_30_date3_meanRelativeHum  # 2 hour ahead of UTC time (summer)

climateHour_date4_meanRelativeHum = climateHour[climateHour["from"] == "2022/10/30 03:00:00"]["mean_wind_speed"].reset_index(drop=True)[0]
df_30_date4_meanRelativeHum = df_30[df_30["datetime"] == "2022-10-30 04:00:00"]["mean_wind_speed"].reset_index(drop=True)[0]

climateHour_date4_meanRelativeHum == df_30_date4_meanRelativeHum  # 1 hour ahead of UTC time (winter)
```
